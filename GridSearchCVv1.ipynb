# Grid Search CV

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
print('X shape', X.shape)
print(X)
print('y shape', y.shape)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
# label encoder column 1
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
# label encoder column 2
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
def brain_ann(hidden_layers, kernel_initializer, optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = hidden_layers, kernel_initializer = kernel_initializer, activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = hidden_layers, kernel_initializer = kernel_initializer, activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = kernel_initializer, activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    
    return classifier    
```

```python
classifier = KerasClassifier(build_fn= brain_ann)
parameters = {'batch_size': [10],
              'nb_epoch': [420, 520],
              'hidden_layers': [66, 76],
              'kernel_initializer': ['uniform', 'glorot_uniform'],
              'optimizer': ['adam', 'rmsprop'] }
grid_search = GridSearchCV(estimator=classifier,
                           param_grid=parameters,
                           scoring= 'accuracy',
                           cv=10)

grid_search = grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
best_score = grid_search.best_score_
```
